```python
# 导入相关的库
import requests
import json
import csv
```

# 构造url_list

```python
url_list = []
url_template = 'https://jingzhifunds.hexun.com/jz/JsonData/KaifangPaim.aspx?callback=callback&subtype=1&fundcompany=--%E5%85%A8%E9%83%A8%E5%9F%BA%E9%87%91%E7%AE%A1%E7%90%86%E5%85%AC%E5%8F%B8--&pagesize=20&curpage={num}&sortName=m3&sortType=down&fundisbuy=0'

for num in range(1, 11):
    url = url_template.format(num=num)
    url_list.append(url)

url_list
```

# 函数：请求和提取数据

```python
def get_and_extract_data(url):
    # 发送请求
    response = requests.get(url)
    # 解析网页
    raw_data = response.text
    
    # 提取 callback 函数中的 JSON 数据
    json_data_str = raw_data[len('callback('):-len(')')]
    data = json.loads(json_data_str)

    # 提取 list 中的数据
    fund_list = data['list']
    
    return fund_list
```

# 循环url_list，并将结果写入csv文件

```python
file = open("fund_data_10.csv", mode='a+', newline='', encoding='utf-8')

fieldnames = ['num', 'fundCode', 'fundName', 'fundLink', 'trendLink', 'baLink', 'netValue', 
            'netDate', 'm1', 'm3', 'm6', 'y1', 'y2', 'y3', 'thisyear', 'setup', 'discount', 
            'ratefee', 'buy', 'buyLink', 'dtLink']

writer = csv.DictWriter(file, fieldnames=fieldnames)

# 写入CSV文件的标题行，即字段名列表
writer.writeheader()

for url in url_list:
    print("正在抓取:" + url)
    fund_list = get_and_extract_data(url)
    for fund in fund_list:
        writer.writerow(fund)
    
file.close()
```
