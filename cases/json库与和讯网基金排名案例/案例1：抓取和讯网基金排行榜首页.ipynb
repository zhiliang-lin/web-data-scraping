**目标**：抓取和讯网开放式基金排行榜首页(网址：https://jingzhifunds.hexun.com/fund4th/shouyi.aspx###)的内容，并将结果保存到本地csv文件。

------------------------------ 

**导入相应的库**

```python
import requests
import json
import csv
```

**通过开发者工具找出数据对应的网址(json格式)，并将网址保存为变量`url`**

```python
url = 'https://jingzhifunds.hexun.com/jz/JsonData/KaifangPaim.aspx?callback=callback&subtype=1&fundcompany=--%E5%85%A8%E9%83%A8%E5%9F%BA%E9%87%91%E7%AE%A1%E7%90%86%E5%85%AC%E5%8F%B8--&pagesize=20&curpage=1&sortName=m3&sortType=down&fundisbuy=0'
```

**通过requests库抓取数据，并通过数据处理，将结果保存为一个名为`fund_list`的列表（内含多个字典，每个字典代表一个基金）**

```python
response = requests.get(url)

html = response.text
```

```python
html
```

```python
json_data_str = html[len('callback('):-len(')')]  #  返回的数据格式为 'callback({...})'，这里通过去除 'callback(' 和 ')' 来提取 JSON 数据
# 以上代码等价于：json_data_str = html[9:-1]（字符串索引，从第 9 位开始到倒数第 1 位结束）
```

```python
# test_html = 'callback({"name": "John", "age": 30, "city": "New York"})'
# test_html[9:-1]
```

```python
data = json.loads(json_data_str) # 使用 json.loads() 方法将 字符串转换成 Python 字典
data
```

```python
# 提取 list 中的数据
fund_list = data['list']
fund_list 
```

**将`fund_list`的内容保存为csv文件，并命名为`fund_data.csv`**

```python
# 打开 CSV 文件并写入数据
file = open('fund_data.csv', mode='a+', newline='', encoding='utf-8') 
    # 定义字段名
fieldnames = ['num', 'fundCode', 'fundName', 'fundLink', 'trendLink', 'baLink', 'netValue', 
              'netDate', 'm1', 'm3', 'm6', 'y1', 'y2', 'y3', 'thisyear', 'setup', 'discount', 
              'ratefee', 'buy', 'buyLink', 'dtLink']
    
writer = csv.DictWriter(file, fieldnames=fieldnames)
    
# 写入 CSV 文件头
writer.writeheader()
    
# 写入每一条基金数据
for fund in fund_list:
    writer.writerow(fund)

file.close()
```
