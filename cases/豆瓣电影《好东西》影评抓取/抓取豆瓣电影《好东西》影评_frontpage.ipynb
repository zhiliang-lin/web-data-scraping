# 第一部分：利用Selenium获取cookie信息

```python
from selenium import webdriver

# 启动浏览器并登录
driver = webdriver.Firefox()
driver.get("https://www.douban.com")

# 等待网页加载并模拟登录
```

```python
# 在登录后获取所有 Cookie
cookies = driver.get_cookies()
cookies
```

```python
# 打印所有 Cookie
for cookie in cookies:
    print(cookie)

driver.quit() # 关闭浏览器
```

```python
# 将Selenium的cookies转化为requests可以使用的格式
cookie_dict = {cookie['name']: cookie['value'] for cookie in cookies}

cookie_dict
```

```python
cookie_string = "; ".join([f"{key}={value}" for key, value in cookie_dict.items()])

cookie_string
```

# 第四部分：提取数据（整合成一个csv文件）

整合代码，将结果保存成一个csv文件

```python
from pyquery import PyQuery as pq

doc = pq(html)

review_items = doc('div.review-item')


review_list =[]

for review in review_items.items():
    title = review("div.main-bd h2 a").text()
    title_link = review("div.main-bd h2 a").attr("href")
    useful_count = review('a.action-btn.up span').text()
    useless_count = review('a.action-btn.down span').text()
    name = review('a.name').text()
    name_link = review('a.name').attr("href")
    time = review('span.main-meta').text()
    rating = review('span.main-title-rating').attr('title')
    

    review_dict = {
        "title": title,
        "title_link": title_link,
        "useful_count": useful_count,
        "useless_count": useless_count,
        "name": name,
        "name_link": name_link,
        "time": time,
        "rating": rating
    }
    
    
    review_list.append(review_dict)
```

```python
review_list
```

```python
review_list[0].keys()
```

```python
import csv

# 打开文件
file = open("herstory_reviews.csv", 'a+', encoding='utf-8', newline='')     
fieldnames = ['title', 'title_link', 'useful_count', 'useless_count', 'name', 'name_link', 'time', 'rating']  
writer = csv.DictWriter(file, fieldnames=fieldnames)    
writer.writeheader() 

for review in review_list:
    writer.writerow(review)

file.close()
```

```python
import requests
from pyquery import PyQuery as pq
import csv
import time
import random
```

```python
url_list = []
url_template = "https://movie.douban.com/subject/36154853/reviews?start={page}"


for p in range(1, 277): # 共有276页
    url = url_template.format(page=(p-1)*20)
    url_list.append(url)
    
url_list
    
```

```python
def get_html(url):

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
                AppleWebKit/537.36 (KHTML, like Gecko) \
               Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0',
               'cookie': 'll="118282"; bid=G48JUWlmCDU; _pk_id.100001.8cb4=c2ada79b5f9689d2.1734136691.; __utmc=30149280; __utmt=1; dbcl2="236694716:SK69QfEZomw"; ck=Pllz; _pk_ses.100001.8cb4=1; ap_v=0,6.0; push_noty_num=0; push_doumail_num=0; __utma=30149280.24898555.1734136692.1734136692.1734136692.1; __utmz=30149280.1734136692.1.1.utmcsr=(direct)|utmccn=(direct)|utmcmd=(none); __utmv=30149280.23669; __utmb=30149280.3.10.1734136692; __yadk_uid=hgLT9tZx1XhZwkmuMpVdcA03idVTdBVp' #如果是通过开发者工具获取cookie则直接改变这里
               }

    # 发送带 Cookie 的请求
    response = requests.get(url, headers=headers)
    html = response.text
    return html
```

```python
def extract_review_list(html):

    doc = pq(html)

    review_items = doc('div.review-item')

    review_list =[]

    for review in review_items.items():
        title = review("div.main-bd h2 a").text()
        title_link = review("div.main-bd h2 a").attr("href")
        useful_count = review('a.action-btn.up span').text()
        useless_count = review('a.action-btn.down span').text()
        name = review('a.name').text()
        name_link = review('a.name').attr("href")
        time = review('span.main-meta').text()
        rating = review('span.main-title-rating').attr('title')
    

        review_dict = {
            "title": title,
            "title_link": title_link,
            "useful_count": useful_count,
            "useless_count": useless_count,
            "name": name,
            "name_link": name_link,
            "time": time,
            "rating": rating
        }
    
    
        review_list.append(review_dict)
        
    return review_list
```

```python
# 打开文件
file = open("herstory_reviews_frontpage_276.csv", 'a+', encoding='utf-8', newline='')     
fieldnames = ['title', 'title_link', 'useful_count', 'useless_count', 'name', 'name_link', 'time', 'rating']   
writer = csv.DictWriter(file, fieldnames=fieldnames)    
writer.writeheader() 

for url in url_list:
    
    print(f"正在抓取：{url}")
    html = get_html(url)
    
    wait_time = random.randint(23, 35) # 生成 23 到 25 秒的随机浮点数
    print(f"等待{wait_time}秒")
    
    time.sleep(wait_time)   # 暂停指定的秒数
    
    review_list = extract_review_list(html)
    for review in review_list:
        writer.writerow(review)


print("抓取完成")
    
file.close()
```
