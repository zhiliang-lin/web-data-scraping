```python
from pyquery import PyQuery as pq

def extract_review(html):
    doc = pq(html)
    
    content = doc('div.review-content.clearfix').text()
    ip_addr = doc("div.main-meta span:nth-child(3)").text()
    title_link = doc("div.action-collect a").attr("data-url")
    
    my_dict = {
        'content': content,
        'ip_addr': ip_addr,
        'title_link': title_link
    }
        
    return my_dict    

```

```python
import csv
import random
import time

# 打开文件
file = open("herstory_reviews_innerpage.csv", 'a+', encoding='utf-8', newline='')     
fieldnames = ['content', 'ip_addr','title_link']  
writer = csv.DictWriter(file, fieldnames=fieldnames)    
writer.writeheader() 

for url in url_list:
    print(f"正在采集:{url}")
    
    html = get_html(url)
    
    wait_time = random.randint(30, 55) #生成10~15秒的随机等待时间
    print(f"等待{wait_time}秒")
    
    time.sleep(wait_time)
    
    
    my_dict = extract_review(html)
      
    writer.writerow(my_dict)

print("采集完成")
    
file.close()
```
