```python
import requests
from pyquery import PyQuery as pq
import csv
import time
import random
```

```python
url_template = 'https://movie.douban.com/subject/36154853/comments?start={num}&limit=20&status=P&sort=new_score'

url_list = []

for page in range(1, 21):
    url = url_template.format(num=20 * (page - 1))

    url_list.append(url)

url_list
    
```

```python
def get_html(url):

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
                AppleWebKit/537.36 (KHTML, like Gecko) \
               Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0',
               'cookie': 'll="118282"; bid=G48JUWlmCDU; _pk_id.100001.8cb4=c2ada79b5f9689d2.1734136691.; __utmc=30149280; __utmt=1; dbcl2="236694716:SK69QfEZomw"; ck=Pllz; _pk_ses.100001.8cb4=1; ap_v=0,6.0; push_noty_num=0; push_doumail_num=0; __utma=30149280.24898555.1734136692.1734136692.1734136692.1; __utmz=30149280.1734136692.1.1.utmcsr=(direct)|utmccn=(direct)|utmcmd=(none); __utmv=30149280.23669; __utmb=30149280.3.10.1734136692; __yadk_uid=hgLT9tZx1XhZwkmuMpVdcA03idVTdBVp' #如果是通过开发者工具获取cookie则直接改变这里
               }

    # 发送带 Cookie 的请求
    response = requests.get(url, headers=headers)
    html = response.text
    return html
```

```python
from pyquery import PyQuery as pq

def extract_comment_list(html):

    doc = pq(html)

    comment_items = doc('div.comment-item')


    comment_list =[]

    for item in comment_items.items():
        comment_text = item('span.short').text()
        id_name = item('span.comment-info a').text()
        user_id_link = item('span.comment-info a').attr("href")
        rating = item("span.comment-info span:nth-child(3)").attr("title")
    
        comment_dict = {
            'id_name': id_name,
            'comment_text': comment_text,
            'user_id_link': user_id_link,
            'rating': rating
        }   
    
    
        comment_list.append(comment_dict)
        
    return comment_list
```

------------------------------ 

```python
# 打开文件
file = open("herstory_hot_comments.csv", 'a+', encoding='utf-8', newline='')     
fieldnames = ['id_name', 'comment_text', 'user_id_link', 'rating']  
writer = csv.DictWriter(file, fieldnames=fieldnames)    
writer.writeheader() 

for url in url_list:
    
    print(f"正在抓取：{url}")
    html = get_html(url)
    
    wait_time = random.randint(3, 5) # 生成 3 到 5 秒的随机浮点数
    print(f"等待{wait_time}秒")
    
    time.sleep(wait_time)   # 暂停指定的秒数
    
    comment_list = extract_comment_list(html)
    for comment in comment_list:
        writer.writerow(comment)


print("抓取完成")
    
file.close()
```
