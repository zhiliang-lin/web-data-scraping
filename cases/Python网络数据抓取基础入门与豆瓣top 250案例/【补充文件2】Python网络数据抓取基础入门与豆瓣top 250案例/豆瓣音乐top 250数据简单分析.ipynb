# 多变量分析

## 分组比较

```python
df.groupby(by="musician").score.mean() # 不同作者书籍的平均评分
```

```python
pd.pivot_table(df,index=["type"],values=["score","people_num"],aggfunc="mean")
```

```python
filtered_musician_df.groupby(by="musician").score.mean() # 不同音乐人的平均评分
```

```python
pd.pivot_table(filtered_musician_df ,index=["musician"],values=["score","people_num"],aggfunc="mean")
```
