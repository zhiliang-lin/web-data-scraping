# requests + pyquery抓取毒鸡汤

```python
import requests
from pyquery import PyQuery as pq
```

```python
url = 'http://www.nows.fun/'
response = requests.get(url)
```

```python
html = response.content.decode('utf-8')
print(html)
```

```python
doc = pq(html)
text = doc('span#sentence').text()
print(text)
```
