# 案例2：在百度新闻（news.baidu.com）搜索“小米”，获取前10页新闻标题

```python
#函数：get_html(url)
#获取网页的 HTML 内容
def get_html(url):
    browser.get(url)
    html = browser.page_source
    return html
```

```python
template = 'https://www.baidu.com/s?ie=utf-8&medium=0&rtt=1&bsst=1&rsv_dl=news_b_pn&cl=2&wd=%E5%B0%8F%E7%B1%B3&tn=news&rsv_bp=1&oq=&rsv_btype=t&f=8&x_bfe_rqs=0320800000000000000048&x_bfe_tjscore=0.080000&tngroupname=organic_news&newVideo=12&goods_entry_switch=1&pn={num}'

url_list = []
# 循环 10 页 
for i in range(0, 10):
    url = template.format(num=i*10)
    url_list.append(url)

url_list
```

```python
#函数：get_titles(html)
#获取网页内容中的标题
def get_titles(html):
    doc = pq(html)
    titles = doc('h3')
    return titles
    
```

```python
import csv

# 打开 CSV 文件并创建 CSV 写入器对象
with open('baidu_news_titles_10pages.csv', 'a', newline='', encoding='utf-8') as csvfile:
    # 定义 CSV 写入器
    csv_writer = csv.writer(csvfile)
    
    for url in url_list:
        html = get_html(url)
        print('正在抓取：{url}'.format(url=url))
        titles = get_titles(html)
    
        # 遍历 titles 中的每个标题，并逐行写入 CSV 文件
        for title in titles.items():
            csv_writer.writerow([title.text()])
```
